```python
# Convering non numerical data into numerical
data["Aptitude"] = data["Aptitude"].str.replace("%",'').astype(int)
data["English"] = data["English"].str.replace("%",'').astype(int)
data["Quantitative"] = data["Quantitative"].str.replace("%",'').astype(int)
data["Analytical"] = data["Analytical"].str.replace("%",'').astype(int)
data["Domain"] = data["Domain"].str.replace("%",'').astype(int)
data["Computer Fundamental"] = data["Computer Fundamental"].str.replace("%",'').astype(int)
data["Coding"] = data["Coding"].str.replace("%",'').astype(int)
```

```python
transformedData = pd.get_dummies(data)
```
